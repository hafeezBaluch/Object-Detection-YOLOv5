```python
import torch
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
# Sample Image URL
#BASE_URL = 'https://github.com/ultralytics/yolov5/raw/master/data/images/'
image_url =  'images/cat.jpg'
```

```python
# A batch of images (only one entry here)
imgs = [image_url]

# Inference
results = model(imgs)
```

```python
# Display the results
results.show()
```
